```python
df_FRED.info()
```

```python
df_FRED.head()
```

```python
df_FRED.fillna(method='backfill')
```

```python
corr = df_FRED.corr()
```

```python
idx, vals = [], []
for ix, i in enumerate(corr.columns.values):
    for j in corr.columns.values[ix + 1:]:
        idx.append((i, j))
        vals.append(corr[i][j])

ser = pd.Series(data=vals, index=idx)
ser_ord = ser.sort_values(ascending=False)
ser_ord
```
